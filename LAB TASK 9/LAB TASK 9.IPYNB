```python
df.drop('Age', axis=1, inplace=True) 
```

```python
df.drop('Sex',axis = 1,inplace=True)
```

```python
df['BMI'].mode(0)
```

```python
df['Blood_Pressure_Systolic'].fillna(df['Blood_Pressure_Systolic'].mode(0),inplace=True)
```

```python
def fillnaObjectMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['BMI', 'Blood_Pressure_Systolic', 'Sex']
fillnaObjectMode(columns)

```

```python
def fillnaIntMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['Ethnicity', 'Ethnicity']
fillnaIntMode(columns)

```

```python
def fillnafloatMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['price', 'condition']
fillnafloatMode(columns)
```
