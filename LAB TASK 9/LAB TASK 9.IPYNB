```python
df.drop('date',axis = 1,inplace=True)
```

```python
df.drop('country',axis = 1,inplace=True)
```

```python
df['city'].mode(0)
```

```python
df['city'].fillna(df['city'].mode(0),inplace=True)
```

```python
def fillnaObjectMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['city', 'street', 'statezip']
fillnaObjectMode(columns)

```

```python
def fillnaIntMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['bedroom', 'bathroom']
fillnaIntMode(columns)

```

```python
def fillnafloatMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['price', 'condition']
fillnafloatMode(columns)
```
