```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
X = data.iloc[:, :-1]   # all columns except last
y = data.iloc[:, -1]    # only last column

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
cat_columns = X.select_dtypes(['object']).columns
X[cat_columns] = X[cat_columns].apply(lambda col: pd.factorize(col)[0])

print("Converted object columns:", list(cat_columns))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)
```

```python
BernNB = BernoulliNB()
BernNB.fit(X_train, Y_train)
Y_bpred = BernNB.predict(X_test)

b_accuracy = accuracy_score(Y_test, Y_bpred)
b_precision = metrics.precision_score(Y_test, Y_bpred, average='weighted', labels=np.unique(Y_bpred))
b_recall = metrics.recall_score(Y_test, Y_bpred, average='weighted', labels=np.unique(Y_bpred))
b_f1 = metrics.f1_score(Y_test, Y_bpred, average='weighted', labels=np.unique(Y_bpred))

print("\nBernoulli Naive Bayes")
print("Accuracy:", b_accuracy)
print("Precision:", b_precision)
print("Recall:", b_recall)
print("F1 Score:", b_f1)


# ---------- B. Random Forest ----------
RF = RandomForestClassifier()
RF.fit(X_train, Y_train)
Y_rpred = RF.predict(X_test)

r_accuracy = accuracy_score(Y_test, Y_rpred)
r_precision = metrics.precision_score(Y_test, Y_rpred, average='weighted', labels=np.unique(Y_rpred))
r_recall = metrics.recall_score(Y_test, Y_rpred, average='weighted', labels=np.unique(Y_rpred))
r_f1 = metrics.f1_score(Y_test, Y_rpred, average='weighted', labels=np.unique(Y_rpred))

print("\nRandom Forest")
print("Accuracy:", r_accuracy)
print("Precision:", r_precision)
print("Recall:", r_recall)
print("F1 Score:", r_f1)


# ---------- C. Gaussian NB ----------
GausNB = GaussianNB()
GausNB.fit(X_train, Y_train)
Y_gpred = GausNB.predict(X_test)

g_accuracy = accuracy_score(Y_test, Y_gpred)
g_precision = metrics.precision_score(Y_test, Y_gpred, average='weighted', labels=np.unique(Y_gpred))
g_recall = metrics.recall_score(Y_test, Y_gpred, average='weighted', labels=np.unique(Y_gpred))
g_f1 = metrics.f1_score(Y_test, Y_gpred, average='weighted', labels=np.unique(Y_gpred))

print("\nGaussian Naive Bayes")
print("Accuracy:", g_accuracy)
print("Precision:", g_precision)
print("Recall:", g_recall)
print("F1 Score:", g_f1)
```

```python
# ---------- D. Decision Tree ----------
Dtree = DecisionTreeClassifier()
Dtree.fit(X_train, Y_train)
Y_dpred = Dtree.predict(X_test)

d_accuracy = accuracy_score(Y_test, Y_dpred)
d_precision = metrics.precision_score(Y_test, Y_dpred, average='weighted', labels=np.unique(Y_dpred))
d_recall = metrics.recall_score(Y_test, Y_dpred, average='weighted', labels=np.unique(Y_dpred))
d_f1 = metrics.f1_score(Y_test, Y_dpred, average='weighted', labels=np.unique(Y_dpred))

print("\nDecision Tree")
print("Accuracy:", d_accuracy)
print("Precision:", d_precision)
print("Recall:", d_recall)
print("F1 Score:", d_f1)


# ---------- E. Multinomial NB ----------
MultiNB = MultinomialNB()
MultiNB.fit(X_train, Y_train)
Y_mpred = MultiNB.predict(X_test)

m_accuracy = accuracy_score(Y_test, Y_mpred)
m_precision = metrics.precision_score(Y_test, Y_mpred, average='weighted', labels=np.unique(Y_mpred))
m_recall = metrics.recall_score(Y_test, Y_mpred, average='weighted', labels=np.unique(Y_mpred))
m_f1 = metrics.f1_score(Y_test, Y_mpred, average='weighted', labels=np.unique(Y_mpred))

print("\nMultinomial Naive Bayes")
print("Accuracy:", m_accuracy)
print("Precision:", m_precision)
print("Recall:", m_recall)
print("F1 Score:", m_f1)
```

```python
KNN = KNeighborsClassifier()
KNN.fit(X_train, Y_train)
Y_kpred = KNN.predict(X_test)

k_accuracy = accuracy_score(Y_test, Y_kpred)
k_precision = metrics.precision_score(Y_test, Y_kpred, average='weighted', labels=np.unique(Y_kpred))
k_recall = metrics.recall_score(Y_test, Y_kpred, average='weighted', labels=np.unique(Y_kpred))
k_f1 = metrics.f1_score(Y_test, Y_kpred, average='weighted', labels=np.unique(Y_kpred))

print("\nK-Nearest Neighbors")
print("Accuracy:", k_accuracy)
print("Precision:", k_precision)
print("Recall:", k_recall)
print("F1 Score:", k_f1)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("data.csv")
```

```python
# Fill missing values
data = data.fillna(data.mode().iloc[0])

# Convert object â†’ int
cat_cols = data.select_dtypes(include=['object']).columns
data[cat_cols] = data[cat_cols].apply(lambda col: pd.factorize(col)[0])

# Split x & y
x = data.iloc[:, 0:-1]
y = data.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
b_pred = bnb.predict(x_test)

b_accuracy = accuracy_score(y_test, b_pred)
b_precision = precision_score(y_test, b_pred, zero_division=0)
b_recall = recall_score(y_test, b_pred, zero_division=0)
b_f1 = f1_score(y_test, b_pred, zero_division=0)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
r_pred = rf.predict(x_test)

r_accuracy = accuracy_score(y_test, r_pred)
r_precision = precision_score(y_test, r_pred, zero_division=0)
r_recall = recall_score(y_test, r_pred, zero_division=0)
r_f1 = f1_score(y_test, r_pred, zero_division=0)
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
g_pred = gnb.predict(x_test)

g_accuracy = accuracy_score(y_test, g_pred)
g_precision = precision_score(y_test, g_pred, zero_division=0)
g_recall = recall_score(y_test, g_pred, zero_division=0)
g_f1 = f1_score(y_test, g_pred, zero_division=0)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
d_pred = dt.predict(x_test)

d_accuracy = accuracy_score(y_test, d_pred)
d_precision = precision_score(y_test, d_pred, zero_division=0)
d_recall = recall_score(y_test, d_pred, zero_division=0)
d_f1 = f1_score(y_test, d_pred, zero_division=0)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
m_pred = mnb.predict(x_test)

m_accuracy = accuracy_score(y_test, m_pred)
m_precision = precision_score(y_test, m_pred, zero_division=0)
m_recall = recall_score(y_test, m_pred, zero_division=0)
m_f1 = f1_score(y_test, m_pred, zero_division=0)
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
k_pred = knn.predict(x_test)

k_accuracy = accuracy_score(y_test, k_pred)
k_precision = precision_score(y_test, k_pred, zero_division=0)
k_recall = recall_score(y_test, k_pred, zero_division=0)
k_f1 = f1_score(y_test, k_pred, zero_division=0)
```

```python
plt.figure(figsize=(16,10))

left = [1, 2, 3, 4, 5, 6]
height = [b_f1, r_f1, g_f1, d_f1, m_f1, k_f1]
tick_label = ['Bernoulli', 'Random Forest', 'Gaussian', 'Decision Tree', 'Multinomial', 'KNeighbors']

plt.bar(left, height, tick_label=tick_label, width=0.9)
plt.xlabel("Classifiers")
plt.ylabel("F1 Score")
plt.title("F1 Scores of Applied Classifiers")
plt.show()
```
